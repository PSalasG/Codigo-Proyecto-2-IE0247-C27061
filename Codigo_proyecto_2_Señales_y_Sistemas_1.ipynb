Se importan las librerías necesarias y se crear todos los parámetros necesarios para generar las señales.

```python
import numpy as np
import plotly.graph_objects as go
from scipy.signal import butter, filtfilt

T = 5.0
fs = 30.0
cutoff = 2
nyq = 0.5 * fs
order = 2
n = int(T * fs)

# Generar el vector de tiempo 't'
t = np.linspace(0, T, n, endpoint=False)
```

Se crea la señal senoidal y se le agrega ruido.

```python
# Onda senoidal
sig = np.sin(1.2 * 2 * np.pi * t)

# Agregar ruido
noise = 1.5 * np.cos(9 * 2 * np.pi * t) + 0.5 * np.sin(12.0 * 2 * np.pi * t)
data = sig + noise
```

Se define la función que realiza el filtro pasa bajo a la señal.

```python
def butter_lowpass_filter(data, cutoff, fs, order):
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = filtfilt(b, a, data)
    return y
```

Se realiza el filtrado a la señal con ruido y se crean los espectros de la señal original y la filtrada.

```python
# Filtrar los datos
filtered_data = butter_lowpass_filter(data, cutoff, fs, order)

# Calcular los espectros de las señales
espectro_data = np.fft.fft(data)
espectro_filtrado = np.fft.fft(filtered_data)
frecuencias = np.fft.fftfreq(len(data), 1/fs)
```

# Graficación de las señales
Se crea y se muestra la gráfica de la señal senoidal con ruido y la señal filtrada.

```python
# Crear figura para las formas de onda
fig_waveforms = go.Figure()
fig_waveforms.add_trace(go.Scatter(x=t, y=data, line=dict(shape='spline'), name='Señal con ruido'))
fig_waveforms.add_trace(go.Scatter(x=t, y=filtered_data, line=dict(shape='spline'), name='Señal filtrada'))

# Establecer etiquetas y títulos para la figura de formas de onda
fig_waveforms.update_layout(
    title='Formas de Onda antes y después del Filtrado',
    xaxis_title='Tiempo',
    yaxis_title='Amplitud'
)

# Mostrar la figura de formas de onda
fig_waveforms.show()
```

Se crea y se muestra la gráfica del espectro de la señal original.

```python
# Crear figura para el espectro original
fig_original_spectrum = go.Figure()
fig_original_spectrum.add_trace(go.Scatter(x=frecuencias, y=np.abs(espectro_data), mode='lines', name='Espectro Original'))

# Establecer etiquetas y títulos para la figura del espectro original
fig_original_spectrum.update_layout(
    title='Espectro de Frecuencia Original',
    xaxis_title='Frecuencia (Hz)',
    yaxis_title='Amplitud'
)

# Mostrar la figura del espectro original
fig_original_spectrum.show()
```

Se crea y se muestra la gráfica del espectro de la señal filtrada.

```python
# Crear figura para el espectro filtrado
fig_filtered_spectrum = go.Figure()
fig_filtered_spectrum.add_trace(go.Scatter(x=frecuencias, y=np.abs(espectro_filtrado), mode='lines', name='Espectro Filtrado'))

# Establecer etiquetas y títulos para la figura del espectro filtrado
fig_filtered_spectrum.update_layout(
    title='Espectro de Frecuencia Filtrado',
    xaxis_title='Frecuencia (Hz)',
    yaxis_title='Amplitud'
)

# Mostrar la figura del espectro filtrado
fig_filtered_spectrum.show()
```
